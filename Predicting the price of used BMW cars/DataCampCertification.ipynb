### Relationship Between Numerical features and the label (Price)

One of the best ways to quickly visualize the relationship between variables is to plot them against each other in a scatter plot. This makes it easy to look for patterns or trends in the data. Here, the numerical features will be plotted against the label - Price to get an idea about the relationship (association) between them. Examining the relationship between the features and the label can provide key insights to our data. Here all the numerical features will be plotted against the price and the correlation metric returned for each pairing with the price column. The correlation measures the strength of a linear relationship, and ranges from -1 to 1. Highly associated variables with a positive linear relationship will have a correlation close to 1. Highly associated variables with a negative linear relationship will have a linear association close to -1. Variables that do not have a linear association will have correlations close to 0.

```python
for col in numeric_features:
    fig = plt.figure(figsize = (9, 6))
    ax = fig.gca()
    feature = bmw_cars[col]
    label = bmw_cars['price']
    correlation = feature.corr(label)
    sns.regplot(x = feature, y = label, marker = 'o', scatter_kws={'alpha':0.2,'s':2})
    plt.xlabel(col)
    plt.ylabel('Price')
    ax.set_title('Price vs ' + col + ' - correlation: ' + str(correlation))
plt.show()
```

The results from the above visualizations are not conclusive. The price vs tax diagram shows a slight linear relationship between the variables with a correlation of 0.2. 

There seems to be an inverse relationship between the price and mpg variables. Thus, higher values of mpg coincide with lower prices. But this relationship is not very strong as evident from the correlation coefficient of -0.2.

Meanwhile, the price and mileage can be seen to show a negative correlation. Higher mileage values tend to coincide with lower prices. This could mean that bmw cars with a higher mileage sell for cheaper prices than the those with lower mileages.

#### Evaluating hypotheses with Linear Regression Model
The hypotheses stated above would be evaluated one-by-one using Linear Regression models. 
The very first thing to keep in mind is that correlation is not causation. Linear Regression helps to understand the relationships between the variables better, and it may lead to finding that some of the variables explain a large portion of the changes in car prices.

```python
import statsmodels.formula.api as smf

results = smf.ols('price ~ mileage', data = bmw_cars).fit()
print(results.params)
print(results.summary())
```

#### Explanation of the above results
P-Value: The p-value here denotes the probability of getting a similar change in `price` even when there is no relationship between `mileage` and `price`. If the probability is small enough (a very well known threshold is 5%), then it means there has to be some degree of `mileage` effect on the changes in `price`. The p-value here is zero, this means that assuming `mileage` does not have any effect on the changes in `price`, the chance of getting a similar change in `price` is 0%. Observing such a rare event with this data, it does not make sense to believe that `mileage` does not have anything to do with the `price`. It can therefore be concluded that `mileage` has something to do with `price`.

coef: The coefficient can be interpreted as, how much of the change in `price` can be influenced by one value increase in `mileage`. It is -0.2752, meaning that as the `mileage` of the car increases by 1km, the price drops by -0.2752 dollars. 

R-squared: This is used to measure how much of the variability of the `price` this model can explain. The values of R-squared varies between 0 and 1. 1 is the highest, meaning the model can explain 100% of the variability of the target variable.

Root Mean Squared Error (RMSE): The rmse shows the average difference between the actual values and the values this model would predict. Here it is about 9095.74 as calculated in the next cell. This means that the values predicted by this model make an error of 9095.74 dollars.

In conclusion, the `mileage` of a car has an effect on the price, an increase of 1km in `mileage` leads to a -0.2752 dollars decrease in `price`.

### Model Building

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor, RandomForestRegressor
```

```python
#Separating the data into features and labels
X = bmw_cars[['model', 'year', 'transmission', 'fuelType', 'engineSize', 'mileage', 'tax', 'mpg']].values
y = bmw_cars['price'].values
```

```python
# Split data 70%-30% into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)

print ('Training Set: %d rows\nTest Set: %d rows' % (X_train.shape[0], X_test.shape[0]))
```

```python
#Define preprocessing for numeric columns (scale them)

numeric_features = [5, 6, 7]
numeric_transformer = Pipeline(steps = [
    ('scaler', StandardScaler())
])

#Define preprocessing for categorical features (encode them)

categorical_features = [0, 1, 2, 3, 4]
categorical_transformer = Pipeline(steps = [
    ('onehot', OneHotEncoder(handle_unknown = 'ignore'))
]) 

#Combine preprocessing steps
preprocessor = ColumnTransformer(
    transformers = [
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)])

#Creating preprocessing and training pipeline
pipeline = Pipeline(steps = [('preprocessor', preprocessor),
                             ('regressor', GradientBoostingRegressor())])


#Fit the pipeline to train a linear regression model on the training set
model = pipeline.fit(X_train, (y_train))
print(model)
```

```python
#Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print('MSE:', mse)
rmse = np.sqrt(mse)

print('RMSE:', rmse)
r2 = r2_score(y_test, predictions)
print('R2:', r2)


#Plot the predicted vs actual

plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Price of cars')

z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test, p(y_test), color = 'magenta')
plt.show()
```

```python
#Using a different estimator in the pipeline
pipeline = Pipeline(steps = [('preprocessor', preprocessor),
                             ('regressor', RandomForestRegressor())])

#Fit the pipeline to train a linear regression model on the training set
model = pipeline.fit(X_train, (y_train))
print(model)
```

```python
#Evaluate the model using the test data
predictions = model.predict(X_test)
mse = mean_squared_error(y_test, predictions)
print('MSE:', mse)
rmse = np.sqrt(mse)

print('RMSE:', rmse)
r2 = r2_score(y_test, predictions)
print('R2:', r2)


#Plot the predicted vs actual

plt.scatter(y_test, predictions)
plt.xlabel('Actual Labels')
plt.ylabel('Predicted Labels')
plt.title('Price of cars')

z = np.polyfit(y_test, predictions, 1)
p = np.poly1d(z)
plt.plot(y_test, p(y_test), color = 'magenta')
plt.show()
```

There's a definite diagonal trend, and the intersections of the predicted and actual values are generally following the path of the trend line; but there's a fair amount of difference between the ideal function represented by the line and the results. This variance represents the *residuals* of the model - in other words, the difference between the label predicted when the model applies the coefficients it learned during training to the validation data, and the actual value of the validation label. These residuals when evaluated from the validation data indicate the expected level of *error* when the model is used with new data for which the label is unknown.

You can quantify the residuals by calculating a number of commonly used evaluation metrics. We'll focus on the following three:

- **Mean Square Error (MSE)**: The mean of the squared differences between predicted and actual values. This yields a relative metric in which the smaller the value, the better the fit of the model
- **Root Mean Square Error (RMSE)**: The square root of the MSE. This yields an absolute metric in the same unit as the label (in this case, numbers of rentals). The smaller the value, the better the model (in a simplistic sense, it represents the average number of rentals by which the predictions are wrong!)
- **Coefficient of Determination (usually known as *R-squared* or R<sup>2</sup>**): A relative metric in which the higher the value, the better the fit of the model. In essence, this metric represents how much of the variance between predicted and actual label values the model is able to explain.
